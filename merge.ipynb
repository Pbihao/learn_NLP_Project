```python
import pandas as pd
f =  open("./data/training_set.txt")
label = []
text = []
x = f.readline()
while x:
    (lab, txt) = x.split('\t')
    txt = txt[:-2]
    label.append(lab)
    text.append(txt)
    x = f.readline()
label = list(map(lambda x: 'negative' if x == '1' else 'positive', label))

add_data = pd.DataFrame({'text':text, 'label': label})
add_data
```

```python

train_data = pd.read_csv('./data/train_data.csv', index_col=0)
train_data
```

```python
merge_data = train_data.append(add_data, ignore_index=True)

merge_data.to_csv("./data/merge.csv")
merge_data
```
